Two datasets can have the same mean but very different spreads.

## Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

ANS. Skewness measures the asymmetry of a data distribution around its mean. In a perfectly symmetrical distribution, like the normal (bell curve), the mean, median, and mode are equal, and skewness = 0.

But in real-world data, distributions are often skewed‚Äîthey have a longer tail on one side.

 Types of Skewness

1Ô∏è‚É£ Positive Skewness (Right Skewed)

Tail on the right side (higher values stretch out)

Mean > Median > Mode

A few high values pull the mean upward

2Ô∏è‚É£ Negative Skewness (Left Skewed)

Tail on the left side (lower values stretch out)

Mean < Median < Mode

A few very low values pull the mean downward


3Ô∏è‚É£ Zero Skewness (Symmetrical Distribution)

Mean = Median = Mode

The data is evenly distributed

Shape is normal bell curve
